```python
# Create columns that will distinguish defoliated from non defoliated.
```

```python
# import libraries
import geopandas as gp
from pathlib import Path
import numpy as np
```

```python
# read in files

outfolder = Path("/Users/jgoldman/Library/CloudStorage/OneDrive-UniversityofToronto/Data/fire-only-sbw-only-perimeters/cleaned")
defolFire = gp.read_file(outfolder/ 'defoliated-compiled.shp')
NdefolFire = gp.read_file(outfolder/ 'non-defoliated-compiled.shp')
```

```python
# add a defolPresence column with value of 1 for non-defoliated
add_fields = ['defolPresence']
field_name_list = defolFire.columns.tolist()
for fieldToAdd in add_fields:
    if fieldToAdd not in field_name_list:
        #shapefile.reindex(shapefile.columns.tolist().append(fieldToAdd), axis=1)
        defolFire[fieldToAdd]=1
defolFire.head()
```

```python
defolFire.to_file(outfolder / 'defoliated-compiled.shp')
```

```python
# add a defolPresence column with value of 0 for non-defoliated

add_fields = ['defolPresence']
field_name_list = NdefolFire.columns.tolist()
for fieldToAdd in add_fields:
    if fieldToAdd not in field_name_list:
        #shapefile.reindex(shapefile.columns.tolist().append(fieldToAdd), axis=1)
        NdefolFire[fieldToAdd]=0
        

NdefolFire.to_file(outfolder / 'non-defoliated-compiled.shp')
```
