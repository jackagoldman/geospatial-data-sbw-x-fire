```python
data = nbr.merge(defol, on='id', how = 'inner')
print(len(data[data['defol']==1]))
test = data[data['defol']==1]
print(test[test['id']== 11])
dups = test.duplicated()
dups
test[dups]
```

```python
data_slope = data[['id', 'nbr1', 'nbr2', 'nbr3', 
                   'nbr4', 'nbr5', 'nbr6', 'nbr7', 'nbr8', 'nbr9',
                     'nbr10', 'defol', 'tsd', 'years_defol']]
data_slope
```

```python
# write a function that for each fire calculates the slope and outputs as a dataframe
slope = []    
for index,row in data_slope.iterrows():
       df = row[['nbr1', 'nbr2', 'nbr3', 'nbr4', 'nbr5', 'nbr6', 'nbr7', 'nbr8', 'nbr9','nbr10']].values
       print(df)
       result = mk.original_test(df)
       print(result)
       slope.append(result.slope)



data_slope["slope"] = slope  
print(data_slope)



        
```

```python
print(data_slope)
```

```python
defol_slope = data_slope[data_slope['defol']==1]
defol_slope = defol_slope.drop_duplicates(keep ='first')
defol_slope = defol_slope[['slope']].values
non_defol_slope = data_slope[data_slope['defol']==0]
non_defol_slope = non_defol_slope[['slope']].values

stats.ttest_ind(defol_slope, non_defol_slope)
```

```python
data_slope["defol"] = np.where(data_slope["defol"] == 1, "Defoliated", "Not Defoliated")

print(data_slope)

figBoxMed = px.box(data_slope, x="defol", y="nbr1", points = "all",
                   labels={
                     "defol": "Defoliated vs. Non-Defoliated",
                     "slope": "slope of recovery (sen's slope)",
                 },
                  title = "10-year slope of recovery for defolaited vs. non defolaited fires")

figBoxMed.show()
```
