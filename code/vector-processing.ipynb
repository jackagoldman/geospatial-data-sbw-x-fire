```python
# This script processes vector data by compiling individual shapefiles into a single shapefile
# the files that are being created are:
##### 1) defoliated portion of fire perimeter (fire-year defoliated)
##### 2) non-defoliated portion of fire perimeter (fire-year non-defoliated)
```

```python
# import libraries
import geopandas
import os, glob
import pandas
from pathlib import Path

```

```python
# 1) fire-year defoliated
# read in files and concatenate all files

folder = Path("/Users/jgoldman/Library/CloudStorage/OneDrive-UniversityofToronto/Data/fire-only-sbw-only-perimeters/fire-year-defoliated")
outfolder = Path("/Users/jgoldman/Library/CloudStorage/OneDrive-UniversityofToronto/Data/fire-only-sbw-only-perimeters/cleaned")
shapefiles = folder.glob("*.shp")
gdf = pandas.concat([
   geopandas.read_file(str(shp)) # read in all files with the structure .shp
   for shp in shapefiles  # loop - for all shp in shapefiles (folder) pipe to dataframe
]).pipe(geopandas.GeoDataFrame)
gdf.to_file(outfolder / 'compiled.shp') # send it out
```

```python
# 1) fire-year defoliated
# compiled file
defolFire = geopandas.read_file(outfolder/ 'compiled.shp')

defolFire.head()
```

```python
# 2) fire-year non-defoliated
# read in files and concatenate all files

folder = Path("/Users/jgoldman/Library/CloudStorage/OneDrive-UniversityofToronto/Data/fire-only-sbw-only-perimeters/fire-year-non-defoliated")
outfolder = Path("/Users/jgoldman/Library/CloudStorage/OneDrive-UniversityofToronto/Data/fire-only-sbw-only-perimeters/cleaned")
shapefiles = folder.glob("*.shp")
gdf = pandas.concat([
   geopandas.read_file(str(shp)) # read in all files with the structure .shp
   for shp in shapefiles  # loop - for all shp in shapefiles (folder) pipe to dataframe
]).pipe(geopandas.GeoDataFrame)
gdf.to_file(outfolder / 'non-defoliated-compiled.shp') # send it out
```
